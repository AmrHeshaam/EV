# Electric Vehicle Usage for The State of Washington
This notebook explores the dataset of EVs registered through the US state of Washington, aiming to shed light on key insights such as which brands are the most popular.

1. Import Libraries and Load Data:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
ev_data = pd.read_csv('Electric_Vehicle_Population_Data.csv')

# Displaying the first few rows of the dataset to get a basic idea
ev_data.head()
```

2. Data Cleaning:

```python
# Checking for missing values
ev_data.isnull().sum()

# Dropping rows with missing key information (e.g., Make, Model)
ev_data_cleaned = ev_data.dropna(subset=['Make', 'Model', 'City', 'County'])

# Displaying the cleaned data
ev_data_cleaned.info()
```

3. Distribution of EVs by City and County:

```python
# Top 10 Cities with the Most Electric Vehicles
top_cities = ev_data_cleaned['City'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title('Top 10 Cities with the Most Electric Vehicles')
plt.xlabel('Number of Electric Vehicles')
plt.ylabel('City')
plt.show()

# Top 10 Counties with the Most Electric Vehicles
top_counties = ev_data_cleaned['County'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_counties.values, y=top_counties.index, palette='magma')
plt.title('Top 10 Counties with the Most Electric Vehicles')
plt.xlabel('Number of Electric Vehicles')
plt.ylabel('County')
plt.show()
```

4. Popularity of Different EV Brands:

```python
# Most Popular EV Brands
top_brands = ev_data_cleaned['Make'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_brands.values, y=top_brands.index, palette='cool')
plt.title('Top 10 Electric Vehicle Brands in Washington State')
plt.xlabel('Number of Electric Vehicles')
plt.ylabel('Brand')
plt.show()
```

5. Conclusion:
   
    This demo has shown us some insightful information about the details of EVs. We can clearly see that Tesla is by far the most popular brand; and        that King county, and more specifically the city of Seattle, has the highest number of them.
